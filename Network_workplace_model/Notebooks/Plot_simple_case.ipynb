```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df1 = pd.read_csv("simple_Dinit.csv")
IPs = np.sort(np.unique(df1["Infection_Prob"]))
CPs = np.sort(np.unique(df1["Contact_Prob"]))
dfh = df1[(df1["Infection_Prob"]==IPs[2])&(df1["Contact_Prob"]==CPs[2])]
figh = sns.catplot(x="Isolation_Prob", y="Frac_recovered", hue="Group", col="Driver_Mixing_Time",\
            row="Mixing_Param", data=dfh, kind="box")
figh.savefig("Driver_init_ip%d_cp%d.png"%(int(100*IPs[1]),int(100*CPs[2])))
```
