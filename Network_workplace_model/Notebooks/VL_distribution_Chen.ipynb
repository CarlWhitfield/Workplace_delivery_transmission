```julia
using QuadGK

sens_max, disp, VL0, wb_shape, wb_scale = 0.91, 3.0, 1.5, 3.5, 7.0
function PCR_positive_dist(x)
    return sens_max*pdf(Weibull(wb_shape, wb_scale),x)/(1 + exp(-disp*(x-VL0)))
end


print(quadgk(PCR_positive_dist,0,15)[1])
x1 = 0:0.01:12
```

```julia
sens_max, disp, VL0, wb_shape, wb_scale = 0.89, 0.5, 4.5, 3.5, 7.0
function LFD_positive_dist(x)
    VLpdf = pdf(Weibull(wb_shape, wb_scale),x)
    av_infectivity = 1.3 .^(x-6.3)
    ai0 = 1.3 .^(VL0-6.3)
    return sens_max*VLpdf ./ (1 .+ exp.(-disp .* (x .- VL0)))
end

print(quadgk(LFD_positive_dist,0,15)[1])
```

```julia
using LaTeXStrings

sens_max0, disp0, VL00 = 0.907, 3.0, 1.5

p1 = Plots.plot(x1,sens_max0 ./ (1 .+ exp.(-disp0 .* (x1 .- VL00))),ylim=(0,1),lw=2,
                xlabel=L"\mathrm{Viral} \; \mathrm{load} \; \mathrm{log}_{10} \mathrm{copies/ml}",
                ylabel=L"\mathrm{Probability} \; \mathrm{of} \; \mathrm{detection}",label=L"\mathrm{PCR}")
x2 = 0:0.2:3
sens_max1, disp1, VL01 = 0.89, 0.5, 4.5
V2 = 6.3 .+ log.(x2) ./ log(1.3)
f0 = 1.3 .^(VL01 - 6.3)
p2 = Plots.plot(x2,sens_max1 ./ (1 .+ exp.(-disp1 .* (V2 .- VL01))),ylim=(0,1),color=:red,lw=2,
    xlabel=L"\mathrm{Infectivity} \; \mathrm{(normalised)}",
    ylabel=L"\mathrm{Probability} \; \mathrm{of} \; \mathrm{detection}",label=L"\mathrm{LFD}")
Plots.plot(p1,p2,layout=(1,2),size=(800,400))
Plots.savefig("PCR_vs_LFD_detection.png")
```

```julia
x = 0:40
tsymp = 5
x2 = vcat(-1 .+ (7.3/(tsymp+1)) .* (0:(tsymp+1)), 6.3 .- 0.2.* (1:(40-tsymp-1)))
Plots.plot(x, sens_max0 ./ (1 .+ exp.(-disp0 .* (x2 .- VL00))),xlim=(0,40),lw=2,label=L"\mathrm{PCR}")
beta = 1.0
x3 = cdf(Gamma(tsymp*beta+1,1/beta),x[2:end]) - cdf(Gamma(tsymp*beta+1,1/beta),x[1:(end-1)])
x3 /= mean(x3[cdf(Gamma(tsymp*beta+1,1/beta),x[2:end]) .< 0.99])
V3 = 6.3 .+ log.(x3) ./ log(1.3)
Plots.plot!(x[1:(end-1)] .+ 0.5, sens_max1 ./ (1 .+ exp.(-disp1 .* (V3 .- VL01))),c=:red,lw=2,label=L"\mathrm{LFD}")
Plots.plot!([tsymp+0.5,tsymp+0.5],[0,1],ls=:dash,lc=:grey,xlim=(0,40),ylim=(0,1),
            xlabel=L"\mathrm{Days} \; \mathrm{since} \; \mathrm{infection}",
            ylabel=L"\mathrm{Probability} \; \mathrm{of} \; \mathrm{detection}",
            label=L"\mathrm{Symptom} \; \mathrm{onset}")
Plots.savefig("PCR_vs_LFD_positive_window.png")
```
