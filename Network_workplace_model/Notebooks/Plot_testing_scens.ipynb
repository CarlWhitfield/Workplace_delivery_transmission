```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df1 = pd.read_csv("testing_loops.csv")
dfh = df1[df1["Group"]=="All"]
dfh.insert(0, "TestType", np.empty(len(dfh.index),dtype='object'))
```

```python
dfh.loc[(dfh.protocol == 'LFD_mass_testing'),'TestType'] = "LFD -- 0 delay"
dfh.loc[(dfh.protocol == 'PCR_mass_testing') & (dfh.delay == 0),'TestType'] = "PCR -- 0 delay"
dfh.loc[(dfh.protocol == 'PCR_mass_testing') & (dfh.delay == 1),'TestType'] = "PCR -- 1 day delay"
dfh.loc[(dfh.protocol == 'PCR_mass_testing') & (dfh.delay == 2),'TestType'] = "PCR -- 2 day delay"
```

```python
figh = sns.catplot(x="tperiod", y="IndexCaseInfections", hue="TestType", col="Pisol",\
            row="new_comply_prob", data=dfh, kind="point")
figh.set(ylim=(0,4))
figh.savefig("testing_network_workplace_simple.png")
```

```python
figh = sns.catplot(x="tperiod", y="FracRecovered", hue="TestType", col="Pisol",\
            row="new_comply_prob", data=dfh, kind="point")
figh.savefig("testing_network_workplace_simple_size.png")
```

```python
figh = sns.catplot(x="tperiod", y="FracRecovered", hue="TestType", col="Pisol",\
            row="new_comply_prob", data=dfh, kind="box")
figh.savefig("testing_network_workplace_simple_size.png")
```
