```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
t= np.arange(51)*20
s1 = 0.001
s2 = 0.01
N1 = 7
N2 = 3
med = stats.poisson.median(s2*t/N2)
upp = stats.poisson.ppf(0.975,s2*t/N2)
low = stats.poisson.ppf(0.025,s2*t/N2)
plt.fill_between(t,upp,np.flip(low),color='C0',alpha=0.25)
plt.plot(t,med,label='Every 7 days -- 99.9% spec.')

med = stats.poisson.median(s2*t/N1)
upp = stats.poisson.ppf(0.975,s2*t/N1)
low = stats.poisson.ppf(0.025,s2*t/N1)
plt.fill_between(t,upp,np.flip(low),color='C1',alpha=0.25)
plt.plot(t,med,label='Every 7 days -- 99% spec.')

med = stats.poisson.median(s1*t/N2)
upp = stats.poisson.ppf(0.975,s1*t/N2)
low = stats.poisson.ppf(0.025,s1*t/N2)
plt.fill_between(t,upp,np.flip(low),color='C1',alpha=0.25)
plt.plot(t,med,label='Every 3 days -- 99.9% spec.')

med = stats.poisson.median(s1*t/N1)
upp = stats.poisson.ppf(0.975,s1*t/N1)
low = stats.poisson.ppf(0.025,s1*t/N1)
plt.fill_between(t,upp,np.flip(low),color='C2',alpha=0.25)
plt.plot(t,med,label='Every 3 days -- 99% spec.')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Expected number of false positives")
```
